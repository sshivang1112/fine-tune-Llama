```python
import pandas as pd


def security():
    df = pd.read_csv('/Users/ayushtiwari/Desktop/Github Projects/fineTuning/Security_System_data.csv')

    df['timestamp'] = pd.to_datetime(df['timestamp'])


    def label_anomaly(timestamp):
        hour = timestamp.hour
        if 22 <= hour or hour < 5:  
            return 1
        else:
            return 0
    df['anomaly'] = df['timestamp'].apply(label_anomaly)

    df.to_csv('/Users/ayushtiwari/Desktop/Github Projects/fineTuning/Security_System_data.csv', index=False)


def window():
    df = pd.read_csv('window_Shade_data.csv')

    df['timestamp'] = pd.to_datetime(df['timestamp'])


    def label_anomaly(timestamp):
        hour = timestamp.hour
        if 22 <= hour or hour < 5: 
            return 1
        else:
            return 0
    df['anomaly'] = df['timestamp'].apply(label_anomaly)

    df.to_csv('window_Shade_data.csv', index=False)


def Garage():
    df = pd.read_csv('Garage_data.csv')

    df['timestamp'] = pd.to_datetime(df['timestamp'])


    def label_anomaly(timestamp):
        hour = timestamp.hour
        if 22 <= hour or hour < 5:  
            return 1
        else:
            return 0
    df['anomaly'] = df['timestamp'].apply(label_anomaly)

    df.to_csv('Garage_data.csv', index=False)


def lock():
    df = pd.read_csv('Smart_lock_data.csv')

    df['timestamp'] = pd.to_datetime(df['timestamp'])


    def label_anomaly(timestamp):
        hour = timestamp.hour
        if 22 <= hour or hour < 5:  
            return 1
        else:
            return 0
    df['anomaly'] = df['timestamp'].apply(label_anomaly)

    df.to_csv('Smart_lock_data.csv', index=False)


def menu():
    while True:
        print("\nMenu:")
        print("1. Security System")
        print("2. Window Shade")
        print("3.Garage")
        print("4. Smart Lock")
        print("5. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            security()
            print("Security System data processed and saved.")
        elif choice == '2':
            window()
            print("Window Shade data processed and saved.")
        elif choice == '3':
            Garage()
            print("Garage data processed and saved.")
        elif choice == '4':
            lock()
            print("Window Shade data processed and saved.")
        elif choice == '5':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter a valid option.")


menu()
```
